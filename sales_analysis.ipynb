```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

data = {
    "Branch":["A","B","C","A","B"],
    "City":["Yangon","Mandalay","Naypyitaw","Yangon","Mandalay"],
    "Gender":["Male","Female","Male","Female","Male"],
    "Product Line":["Food","Electronics","Sports","Food","Health"],
    "Quantity":[5,3,8,2,6],
    "Total":[500,300,800,200,600],
    "Rating":[8.5,7.2,9.1,6.8,8.0]
}

df = pd.DataFrame(data)

print(df)
```

```python
print("Rows and Columns:", df.shape)
```

```python
print(df.columns)
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
print("Total Sales:", round(df["Total"].sum(),2))
```

```python
print("Average Sales:", round(df["Total"].mean(),2))
```

```python
print(df.loc[df["Total"].idxmax()])
```

```python
print(df.groupby("Branch")["Total"].sum())
```

```python
print(df.groupby("City")["Total"].sum())
```

```python
print(df.groupby("Gender")["Total"].sum())
```

```python
df.groupby("Branch")["Total"].sum().plot(kind="bar")

plt.title("Sales by Branch")
plt.xlabel("Branch")
plt.ylabel("Sales")

plt.show()
```

```python
plt.hist(df["Total"], bins=20)

plt.title("Sales Distribution")
plt.xlabel("Total Sale")
plt.ylabel("Frequency")

plt.show()
```

```python
print("Average Rating:", round(df["Rating"].mean(),2))
```
